### –ó–∞–¥–∞—á–∞ 2. –î–ª—è –∫–∞–∂–¥–æ–π –∫–Ω–∏–≥–∏ –ø–æ—Å—á–∏—Ç–∞–π—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–±–∑–æ—Ä–æ–≤ –∏ —Å—Ä–µ–¥–Ω—é—é –æ—Ü–µ–Ω–∫—É.

```python
query = '''SELECT b.title AS books_title,
                  COUNT(DISTINCT rev.review_id) AS review_count,
                  AVG(rat.rating) AS avg_rating
           FROM books AS b
           LEFT JOIN reviews AS rev ON b.book_id = rev.book_id
           LEFT JOIN ratings AS rat ON b.book_id = rat.book_id
           GROUP BY b.book_id
           ORDER BY avg_rating DESC, review_count DESC
           '''

con=engine.connect()

pd.io.sql.read_sql(sql=text(query), con = con)
```

### –ó–∞–¥–∞—á–∞ 3. –û–ø—Ä–µ–¥–µ–ª–∏—Ç–µ –∏–∑–¥–∞—Ç–µ–ª—å—Å—Ç–≤–æ, –∫–æ—Ç–æ—Ä–æ–µ –≤—ã–ø—É—Å—Ç–∏–ª–æ –Ω–∞–∏–±–æ–ª—å—à–µ–µ —á–∏—Å–ª–æ –∫–Ω–∏–≥ —Ç–æ–ª—â–µ 50 —Å—Ç—Ä–∞–Ω–∏—Ü ‚Äî —Ç–∞–∫ –≤—ã –∏—Å–∫–ª—é—á–∏—Ç–µ –∏–∑ –∞–Ω–∞–ª–∏–∑–∞ –±—Ä–æ—à—é—Ä—ã.

```python
query = '''SELECT publisher
           FROM publishers
           WHERE publisher_id IN (SELECT publisher_id
                                  FROM books
                                  WHERE num_pages > 50
                                  GROUP BY publisher_id
                                  ORDER BY COUNT(book_id) DESC
                                  LIMIT 1)
        '''

con=engine.connect()

pd.io.sql.read_sql(sql=text(query), con = con)
```

### –ó–∞–¥–∞—á–∞ 4. –û–ø—Ä–µ–¥–µ–ª–∏—Ç–µ –∞–≤—Ç–æ—Ä–∞ —Å —Å–∞–º–æ–π –≤—ã—Å–æ–∫–æ–π —Å—Ä–µ–¥–Ω–µ–π –æ—Ü–µ–Ω–∫–æ–π –∫–Ω–∏–≥ ‚Äî —É—á–∏—Ç—ã–≤–∞–π—Ç–µ —Ç–æ–ª—å–∫–æ –∫–Ω–∏–≥–∏ —Å 50 –∏ –±–æ–ª–µ–µ –æ—Ü–µ–Ω–∫–∞–º–∏.

```python
query = '''SELECT author
           FROM authors
           WHERE author_id IN (SELECT author_id
                               FROM books
                               WHERE book_id IN (SELECT book_id
                                                 FROM ratings
                                                 GROUP BY book_id
                                                 HAVING COUNT(rating) >= 50
                                                 ORDER BY AVG(rating) DESC
                                                 LIMIT 1)
                               )
        '''

con=engine.connect()

pd.io.sql.read_sql(sql=text(query), con = con)
```

```python
query = '''SELECT EXTRACT(YEAR FROM publication_date::date) AS publication_year,
                  COUNT(DISTINCT publisher_id) AS publisher_count,
                  COUNT(DISTINCT book_id) AS count_books,
                  ROUND(SUM(num_pages)/1000::NUMERIC, 1) AS pages_sum_thousand
           FROM books
           GROUP BY publication_year
           HAVING COUNT(DISTINCT book_id) > 30
           ORDER BY publication_year
           
        '''

con=engine.connect()

pd.io.sql.read_sql(sql=text(query), con = con)
```

```python
query = '''WITH table_1 AS
                          (SELECT rating AS rating_more_3_users,
                                  book_id
                           FROM ratings
                           WHERE book_id IN (SELECT book_id
                                             FROM reviews
                                             GROUP BY book_id
                                             HAVING COUNT(DISTINCT username) > 3)),
                table_2 AS 
                          (SELECT rating AS rating_less_3_users,
                                  book_id
                           FROM ratings
                           WHERE book_id IN (SELECT book_id
                                             FROM reviews
                                             GROUP BY book_id
                                             HAVING COUNT(DISTINCT username) <= 3))
            SELECT AVG(table_1.rating_more_3_users) AS avg_rating_more,
                   AVG(table_2.rating_less_3_users) AS avg_rating_less
            FROM table_1
            FULL JOIN table_2 ON table_1.book_id = table_2.book_id;
                  
        '''

con=engine.connect()

pd.io.sql.read_sql(sql=text(query), con = con)
```
